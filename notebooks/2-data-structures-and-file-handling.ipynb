### 4. Sets
**Unordered, unique elements**

## File Handling

### 1. Reading Files

### 3. CSV Handling
```python
import csv

# Write CSV
with open("students.csv", "w", newline="") as f:
    writer = csv.writer(f)
    writer.writerow(["Name", "Age", "Major"])
    writer.writerow(["Alice", 25, "Computer Science"])
    writer.writerow(["Bob", 24, "Mathematics"])

# Read CSV
print("CSV contents:")
with open("students.csv", "r") as f:
    reader = csv.reader(f)
    for row in reader:
        print(row)
```

---

````python
import csv
# Write CSV
with open("students.csv", "w", newline="") as f:
    writer = csv.writer(f)
    writer.writerow(["Name", "Age", "Major"])
    writer.writerow(["Alice", 25, "Computer Science"])
    writer.writerow(["Bob", 24, "Mathematics"])
````

````python
# Read CSV
print("CSV contents:")
with open("students.csv", "r") as f:
    reader = csv.reader(f)
    for row in reader:
        print(row)
````
